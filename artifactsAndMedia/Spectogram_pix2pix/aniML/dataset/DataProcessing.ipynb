```python
import cv2 as c
from google.colab.patches import cv2_imshow as cs
```

```python
#Creating our lists and category labels
img = []
sp = []
cat = ['baby', 'cat', 'cow', 'crow', 'dog', 'frog', 'hen', 'pig', 'rooster', 'sheep']

#loading images
for i in cat:
  for j in range(1,11):
    if(j < 10): 
      img.append(c.imread(f'img_{i}_0{j}.png'))
      sp.append(c.imread(f'sp_{i}_0{j}.png'))
      
    if(j == 10): 
      img.append(c.imread(f'img_{i}_{j}.png'))
      sp.append(c.imread(f'sp_{i}_{j}.png'))

#resizing spectograms
for i,o in enumerate(sp):
  sp[i] = c.resize(o, (256,256), interpolation = c.INTER_AREA)

#concatenating images
data =[]
for i in range(100):
  data.append(c.hconcat([img[i],sp[i]]))

#test
cs(data[55])
print(data[23].shape)
```
